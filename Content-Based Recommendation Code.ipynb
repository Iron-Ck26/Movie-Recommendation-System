```python
from sklearn.feature_extraction.text import TfidfVectorizer


tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
```

```python
movies_extract_df = pd.DataFrame(movies_cleaned_df)
```

```python
movies_extract_df.head(10)
```

```python
target_words = ['Action', 'Animation', 'Adventure', 'Fantasy', 'Science Fiction', 'Drama', 'Thriller','Family']
```

```python
def extract_target_words(text, target_words):
    extracted_genres = []
    for target_words in target_words:
        match = re.search(rf'\b{re.escape(target_words)}\b', text, re.IGNORECASE)
        if match:
            extracted_genres.append(match.group())
    return extracted_genres if extracted_genres else None
```

```python
movies_extract_df['extracted_genres'] = movies_extract_df['genres'].apply(lambda x: extract_target_words(x, target_words))
print(movies_extract_df['extracted_genres'])
```

```python
movies_extract_df['overview'] = movies_extract_df['overview'].fillna('NA')
```

```python
movies_extract_df.head(10)
```

```python
# Fitting the TF-IDF on the 'overview' text
tfv_matrix = tfv.fit_transform(movies_extract_df['overview'])
tfv_matrix
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

#Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movies_extract_df.index, index=movies_cleaned_df['original_title']).drop_duplicates()
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))
  

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    recommended_movies = movies_extract_df.iloc[movie_indices][['original_title', 'homepage', 'extracted_genres']]
    return recommended_movies
```
